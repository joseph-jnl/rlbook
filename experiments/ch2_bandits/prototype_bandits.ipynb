```python
import logging
import pandas as pd
import numpy as np
from rlbook.bandits.testbeds import NormalTestbed
from rlbook.bandits.bandits import EpsilonGreedy, init_constant
```

```python
EXPECTED_VALUES = {
    1: {"mean": 2, "var": 1},
    2: {"mean": -1, "var": 1},
    3: {"mean": 1, "var": 1},
    4: {"mean": 0, "var": 1},
    5: {"mean": 1.7, "var": 1},
}

testbed = NormalTestbed(EXPECTED_VALUES)
testbed_drift = NormalTestbed(EXPECTED_VALUES, p_drift=0.0)

bandit = EpsilonGreedy(init_constant(testbed, q_val=10), epsilon=0)
bandit.run(testbed, 2000, n_runs=1)
```
